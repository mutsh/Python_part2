In the code above, the data has been standardized so that the mean is approximately 0 and the standard deviation is close to 1. Standardization is often performed before training machine learning models to ensure that each feature contributes equally to the model. The process involves subtracting the mean from each data point (for that feature) and dividing by the standard deviation:

$$ z = \frac{x - \mu}{\sigma} $$

where:
- *z* is the scaled value.
- *x* is the original value.
- $\mu$ is the mean of the feature.
- $\sigma$ is the standard deviation of the feature.

## Group Exercise

Now, look again at our code from yesterday for making a PCA and producing a biplot. Abstract it where it makes sense and wrap it inside two functions.

The first function should:

* take in the path to/filename of the dataset to read in, a PCA object, and perhaps other parameters you may want to tweak
* return a dataframe of at least the two first PCs and the outcome variable.

Then write a second function that:

* takes in the PCA object, the dataframe returned from the first function, and perhaps other parameters you may want to tweak
* returns the biplot

Here we show one solution. Others are possible, depending on what you find convenient or what you plan to do with the code.

```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
def run_PCA(PCA_obj, filename_csv, numeric_features = [], outcome_variable = ''):

  # Load data
  df = pd.read_csv(filename_csv)

  # Define traning data (exlude outcome/target variable)
  X = df[numeric_features]

  # Scale and center the data
  scaling = StandardScaler()
  X_scale = scaling.fit_transform(X)

  # Fit PCA object to the scaled data
  PCA_obj_fitted = PCA_obj.fit_transform(X_scale)

  #Create column names
  colnames = []
  for i in range(1, PCA_obj_fitted.shape[1]+1):
      colnames.append('PC' + str(i))

  pca_df = pd.DataFrame(PCA_obj_fitted, columns = colnames)

  # Add the target varibale to the PCA-dataframe.
  pca_df['outcome'] = df[outcome_variable]

  return pca_df

def biplot(PCA_obj, pca_df, numeric_features):

  # Scale the first two PC's
  scalePC1 = 2.0/(pca_df['PC1'].max() - pca_df['PC1'].min())
  scalePC2 = 2.0/(pca_df['PC2'].max() - pca_df['PC2'].min())
  pca_df['PC1_scaled'] = pca_df['PC1'] * scalePC1
  pca_df['PC2_scaled'] = pca_df['PC2'] * scalePC2

  # Plot the two first PC (scaled) and color the datapoint by the target value
  plt.figure(figsize=(9,7))

  # Add eigenvectors to the plot to create a biplot
  sns.scatterplot(pca_df, x = 'PC1_scaled', y = 'PC2_scaled', hue = 'outcome')

  for i, feature in enumerate(numeric_features):
      plt.arrow(0, 0,
                PCA_obj.components_[0, i],
                PCA_obj.components_[1, i],
                head_width=0.01,
                head_length=0.02,
                color="gray")
      plt.text(PCA_obj.components_[0, i] * 1.15,
              PCA_obj.components_[1, i] * 1.15,
              feature, fontsize=10)

  return plt
```

```python
# Testing the functions on the clean diabetes dataset
PCA_obj = PCA(n_components = 4)
filename_csv = 'https://raw.githubusercontent.com/Center-for-Health-Data-Science/Python_part2/main/data/diabetes_clean.csv'
numeric_features = ['Age', 'BloodPressure', 'GeneticRisk', 'BMI', 'PhysicalActivity']
outcome_variable = 'Diabetes'

pca_df = run_PCA(PCA_obj = PCA_obj, filename_csv = filename_csv, numeric_features = numeric_features, outcome_variable = outcome_variable)
print(pca_df)
biplot(PCA_obj = PCA_obj, pca_df = pca_df, numeric_features = numeric_features)
```

## Alternative way where we create a PCA object inside function 1

The functions can be made in other ways and take different parameters. This is one suggestion.

```python
def create_PCA_object(filename_csv, numeric_features = [], n_components = 4):

  # Load data
  df = pd.read_csv(filename_csv)

  # Define traning data (exlude outcome/target variable)
  X = df[numeric_features]

  # Scale and center the data
  scaling = StandardScaler()
  X_scale = scaling.fit_transform(X)

  # Define the PCA object
  PCA_obj = PCA(n_components = n_components)

  # Fit PCA object to the scaled data
  PCA_obj_fitted = PCA_obj.fit_transform(X_scale)

  #Create column names
  colnames = []
  for i in range(1, PCA_obj_fitted.shape[1]+1):
      colnames.append('PC' + str(i))

  pca_df = pd.DataFrame(PCA_obj_fitted, columns = colnames)

  return_list = [PCA_obj, pca_df]

  return return_list

def biplot(PCA_obj, pca_df, outcome_variable_array, numeric_features = []):

  # Transform result into dataframe with named columns
  colnames = []
  for i in range(1, pca_df.shape[1]+1):
    colnames.append('PC' + str(i))

  pca_df = pd.DataFrame(pca_df, columns = colnames)

  # Scale the first two PC's
  scalePC1 = 2.0/(pca_df['PC1'].max() - pca_df['PC1'].min())
  scalePC2 = 2.0/(pca_df['PC2'].max() - pca_df['PC2'].min())
  pca_df['PC1_scaled'] = pca_df['PC1'] * scalePC1
  pca_df['PC2_scaled'] = pca_df['PC2'] * scalePC2

  # Add the target varibale to the PCA-dataframe.
  pca_df['outcome'] = outcome_variable_array

  # Plot the two first PC (scaled) and color the datapoint by the target value
  plt.figure(figsize=(9,7))

  # Add egienvectors to the plot to create a biplot
  sns.scatterplot(pca_df, x = 'PC1_scaled', y = 'PC2_scaled', hue = 'outcome')

  for i, feature in enumerate(numeric_features):
      plt.arrow(0, 0,
                PCA_obj.components_[0, i],
                PCA_obj.components_[1, i],
                head_width=0.01,
                head_length=0.02,
                color="gray")
      plt.text(PCA_obj.components_[0, i] * 1.15,
              PCA_obj.components_[1, i] * 1.15,
              feature, fontsize=10)

  return plt
```

## Group Exercise: Data Class

The below is our implementation suggestion. Others are possible.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

class Data:
    def __init__(self, filename):
        """
        Initializes the Data class with a dataset.

        Parameters:
        - filename (str): The path to the file to load
        """
        self.filename = filename
        self.df = None
        self.scaled_data = None
        self.pca_result = None
        self.scaler = None
        self.pca = None
        self.numeric_columns = None

        #run the load data method at initialization
        self.load_data()

    def load_data(self):
        #we should probably wrap try except around this
        self.df = pd.read_csv(self.filename)
        #implement here some checks here if data is loaded correctly

    def set_numeric(self, numeric_columns = 'all'):
        """
        Scales specified numeric columns in the data using standard scaling (zero mean, unit variance).

        Parameters:
        - numeric_columns (list): Optional list of columns to treat as numeric. Default 'all'
        """
        if numeric_columns == 'all':
            self.numeric_columns = self.df.columns
        else:
            # Validate provided numeric columns
            invalid_columns = [col for col in numeric_columns if col not in self.df.columns]
            if invalid_columns:
                raise ValueError(f"Invalid columns specified: {invalid_columns}")

            #if supplied columns are fine save them as numeric
            self.numeric_columns = numeric_columns

    def summary_stats(self):
        """
        Returns summary statistics for numeric columns and unique counts for categorical columns.

        Returns:
        - pd.DataFrame: Summary of numeric and categorical data.
        """
        #check if numeric cols have been defined. If no, refuse
        if not self.numeric_columns:
            print('Please define numeric columns first with .set_numeric()')
            return

        numeric_summary = self.df.describe()
        categorical_summary = self.df.drop(columns=self.numeric_columns).nunique().to_frame(name='Unique Count')
        return numeric_summary, categorical_summary

    def scale(self):
        """
        Scales specified numeric columns in the data using standard scaling (zero mean, unit variance).

        Returns:
        - pd.DataFrame: DataFrame containing the scaled numeric columns and unscaled non-numeric columns.
        """
        #check if numeric cols have been defined. If no, refuse
        if not self.numeric_columns:
            print('Please define numeric columns first with .set_numeric()')
            return

        #define numeric and categorical columns
        numeric_data = self.df[self.numeric_columns]
        non_numeric_data = self.df.drop(columns=self.numeric_columns)

        # Scale the selected numeric columns
        self.scaler = StandardScaler()
        self.scaled_data = self.scaler.fit_transform(numeric_data)
        scaled_numeric_df = pd.DataFrame(self.scaled_data, columns=numeric_data.columns, index=self.df.index)

        # Combine scaled numeric data and non-numeric data
        scaled_df = pd.concat([scaled_numeric_df, non_numeric_data], axis=1)

        return scaled_df

    def correlation_matrix(self, plot = False):
        """
        Computes and optionally plots the correlation matrix for numeric columns.

        Parameters:
        - plot (bool): Whether to display a heatmap of the correlation matrix.

        Returns:
        - pd.DataFrame: Correlation matrix.
        """
        #check if numeric cols have been defined. If no, refuse
        if not self.numeric_columns:
            print('Please define numeric columns first with .set_numeric()')
            return

        corr_matrix = self.df[self.numeric_columns].corr()
        if plot:
            plt.figure(figsize=(10, 8))
            sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r', square=True,
                        center = 0, vmin = -1, vmax = 1)
            plt.title("Correlation Matrix")
            plt.show()
        return corr_matrix

    def plot_column(self, column_name, plot_type = 'histogram'):
        """
        Plots a specified column using the specified plot type.

        Parameters:
        - column_name (str): Name of the column to plot.
        - plot_type (str): Type of plot to use ('histogram', 'boxplot', 'violin').

        Raises:
        - ValueError: If the column name or plot type is invalid.
        """
        # Check for valid column
        if column_name not in self.df.columns:
            raise ValueError(f"Column '{column_name}' does not exist in the data.")

        # Check for valid plot type
        plot_type = plot_type.lower()
        if plot_type not in ['histogram', 'boxplot', 'violin']:
            raise ValueError("Invalid plot type. Choose 'histogram', 'boxplot', or 'violin'.")

        plt.figure(figsize=(8, 6))

        # Generate the plot based on type
        if plot_type == 'histogram':
            plt.hist(self.df[column_name], bins=20, edgecolor='black')
            plt.title(f'Histogram of {column_name}')
            plt.xlabel(column_name)
            plt.ylabel('Frequency')

        elif plot_type == 'boxplot':
            sns.boxplot(y=self.df[column_name])
            plt.title(f'Boxplot of {column_name}')

        elif plot_type == 'violin':
            sns.violinplot(y=self.df[column_name])
            plt.title(f'Violin Plot of {column_name}')

        plt.show()

    def perform_pca(self, n_components: int = 2):
        """
        Performs PCA on the (optionally scaled) data.

        Parameters:
        - n_components (int): Number of principal components to keep.

        Returns:
        - pd.DataFrame: Transformed data in the PCA space.
        """
        if self.scaled_data is None:
            print("Data is not scaled. Please run .scale() first")
            return

        self.pca = PCA(n_components=n_components)
        pca_data = self.pca.fit_transform(self.scaled_data)

        # Naming the columns as PC1, PC2, etc.
        column_names = [f"PC{i+1}" for i in range(n_components)]
        self.pca_result = pd.DataFrame(pca_data, columns=column_names)
        return self.pca_result

    def make_biplot(self):
        """
        Creates a biplot for the PCA results, showing both the scores and the loading vectors.

        Raises:
        - ValueError: If PCA has not been performed.
        """
        if self.pca_result is None or self.pca is None:
            raise ValueError("PCA has not been performed. Call perform_pca first.")

        plt.figure(figsize=(10, 8))

        # Plot the scores (data points in the PCA-transformed space)
        plt.scatter(self.pca_result["PC1"], self.pca_result["PC2"], alpha=0.7)
        plt.xlabel("PC1")
        plt.ylabel("PC2")
        plt.title("PCA Biplot")

        # Plot the loading vectors (original feature contributions)
        loadings = self.pca.components_.T  # Transpose to align with the features
        for i, feature in enumerate(self.numeric_columns):
            plt.arrow(0, 0, loadings[i, 0] * max(self.pca_result["PC1"]),
                      loadings[i, 1] * max(self.pca_result["PC2"]),
                      color='r', alpha=0.7, head_width=0.05)
            plt.text(loadings[i, 0] * max(self.pca_result["PC1"]) * 1.1,
                     loadings[i, 1] * max(self.pca_result["PC2"]) * 1.1,
                     feature, color='g', ha='center', va='center', fontsize=12)

        plt.grid()
        plt.show()
```
